```python
# Batch Gradient Descent:Parameters are updated after computing the gradient of error with respect to the entire training set.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

x = [30, 35, 37, 59, 70, 76, 88, 100 ]
y = [1100, 1423, 1377, 1800, 2304, 2588, 3495, 4839]

x = np.reshape(x,(-1,1))
y = np.reshape(y,(-1,1))

scaler = MinMaxScaler()
scaler.fit(x)
x = scaler.transform(x)

n = len(y)
learningrate = 0.1
```

```python
print("In Batch Gradient Descent Weights are updating after considering all the samples in dataset for finding \nprediction and cost value generation")


theta0 = 0
theta1 = 0

loss = []
iteration = []

for i in range(100):
    
    print("Epoch:",i)
    print("Number of Samples:",len(x))
    
    
    print("\ttheta0:{0} \n\ttheta1:{1}".format(theta0,theta1))

    predicted = theta0 + (theta1 * x)

    cost =(1/n)*np.sum(np.square(y - predicted))
    
    print("\tcost:",cost)
    
    loss.append(cost)
    iteration.append(i)

    dtheta0 = (-(2/n) * np.sum(np.subtract(y,predicted)))
    dtheta1 = (-(2/n) * np.sum(np.multiply(np.subtract(y,predicted),x)))

    theta0 = theta0 - (learningrate * dtheta0)
    theta1 = theta1 - (learningrate * dtheta1)
```

```python
plt.plot(iteration,loss)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid()
```
