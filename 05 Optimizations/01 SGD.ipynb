```python
'''
Stochastic Gradient Descent : Parameters are updated after computing the gradient of error with respect to a single training 
                              example
'''
```

```python
import numpy as np
from numpy import random
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

x = [30, 35, 37, 59, 70, 76, 88, 100]
y = [1100, 1423, 1377, 1800, 2304, 2588, 3495, 4839]

x = np.reshape(x,(-1,1))
y = np.reshape(y,(-1,1))

scaler = MinMaxScaler()
scaler.fit(x)
x = scaler.transform(x)

#scaler.fit(y)
#y = scaler.transform(y
```

```python
theta0 =0
theta1 =0

learningrate = 0.1

loss = []
iteration = []
```

```python
print("In SGD Weights are updated for every samples in dataset also considered singled sample for finding prediction and cost value generation\n")


for steps in range(100):
    
    seed = np.random.random()
    np.random.seed(int(seed))
    np.random.shuffle(x)
    np.random.seed(int(seed))
    np.random.shuffle(y)
    
    for i in range(0,1):
        
        if(steps%25 ==0):
            
            print("Number of Samples:",len(x[i]))
        
            print("\ttheta0:{0} \n\ttheta1:{1}".format(theta0,theta1))

        predicted = theta0 + (x[i] * theta1)
        #print("predicted:=",predicted)

        cost = np.square(np.subtract(y[i],predicted)) / 2
        
        if(steps%25 ==0):
            
            print("\tcost:=",cost)
        
        loss.append(cost)
        iteration.append(steps)

        dtheta0 = -1 * (y[i] - predicted)
        dtheta1 = (-x[i]) * (y[i] - predicted)

        theta0 = theta0 - learningrate * dtheta0
        theta1 = theta1 - learningrate * dtheta1
```

```python
plt.plot(iteration,loss)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid()
```
