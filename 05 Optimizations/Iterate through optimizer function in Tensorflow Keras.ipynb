```python
import tensorflow as tf

tf.enable_eager_execution()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = np.arange(10)
y = 4 * x + 3
noise = np.random.normal(0,1,10)
y = y + noise

fig = plt.figure(figsize=(5,5))
ax = sns.lineplot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
```

```python
M = tf.Variable(5.0) 
C = tf.Variable(0.0) 

loss = []
iteration = []

opt = tf.keras.optimizers.Adam(learning_rate=0.001)

for epoch in range(1000):
    iteration.append(epoch)
    
    opt.minimize(lambda : sum(abs(y - M*x - C)), var_list=[M, C])
    # print(sum((y - M*x - C)**2).numpy())
    loss.append(sum((y - M*x - C)**2).numpy())
    
    
# Function : Y = Mx + C

# M and C parameters

# loss function is lambda: sum(abs(y — M*x — C))

# This translates to the mean absolute error, which is quite similar to the regression model which uses mean squared error

# After 1000 iterations we get the values 4.4131455 and 0.29607454 for M and C. 
# Both plots on the same plane would look like below
```

```python
# Predictd for the final M and C

y_pred = M * x + C
```

```python
plt.plot(iteration,loss)
plt.xlabel("Epoch")
plt.ylabel("Cost")
plt.grid()
```

```python
fig = plt.figure(figsize=(5,5))
ax = sns.lineplot(x, y, label = "Original")
ax = sns.lineplot(x,y_pred,label = "Estimated")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
```

```python
# Example :
# Function : Y = Ax^3 + Bx^2 + Cx + D
# COefficients : A,B,C,D

# predicted = Ax^3 + Bx^2 + Cx + D

# cost = sum(abs(Y -predicted))

x = np.arange(-10, 10)

y = x**3 + x**2 + 5

noise = np.random.normal(0,20,20)
y = y + noise

fig = plt.figure(figsize=(5,5))
ax = sns.lineplot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
```

```python
# Varibales or Coefficients
A = tf.Variable(1.0) 
B = tf.Variable(1.0) 
C = tf.Variable(1.0) 
D = tf.Variable(1.0) 

opt = tf.keras.optimizers.Adam(learning_rate=0.01)

loss_fun = sum(abs(y - A*x**3 + B*x**2 + C*x + D))

print(loss_fun.numpy())


for epoch in range(1000):
    iteration.append(epoch)
    
    opt.minimize(lambda : sum(abs(y - (A*x**3+ B*x**2+C*x+D))),var_list=[A,B,C,D])
    
    loss.append(sum(abs(y - (A*x**3+ B*x**2+C*x+D))).numpy())
    
```

```python
fig = plt.figure(figsize=(5,5))
ax = sns.lineplot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
```

```python
plt.plot(iteration,loss)
plt.xlabel("Epoch")
plt.ylabel("Cost")
plt.grid()
```

```python
# final values for weights

print("A:{0} B:{1} C:{2} and D:{3}".format(A.numpy(),B.numpy(),C.numpy(),D.numpy()))
```

```python
# Get the final outvalue using latest weights

y_pred = A*x**3 + B*x**2 + C*x + D
```

```python
fig = plt.figure(figsize=(5,5))
ax = sns.lineplot(x, y, label = "Original")
ax = sns.lineplot(x,y_pred,label = "Estimated")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
```
