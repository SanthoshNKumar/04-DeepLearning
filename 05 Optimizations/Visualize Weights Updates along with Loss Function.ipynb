```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

x = [30, 35, 37, 59, 70, 76, 88, 100 ]
y = [1100, 1423, 1377, 1800, 2304, 2588, 3495, 4839]

x = np.reshape(x,(-1,1))
y = np.reshape(y,(-1,1))

scaler = MinMaxScaler()
scaler.fit(x)
x = scaler.transform(x)

n = len(y)
learningrate = 0.1
```

```python
theta0 = 0
theta1 = 0

histroy_theta0=[]
histroy_theta1=[]

loss = []
iteration = []

for i in range(100):
    
    iteration.append(i)
    
    predicted = theta0 + (theta1 * x)

    cost =(1/n)*np.sum(np.square(y - predicted))
    
    loss.append(cost)
    

    dtheta0 = (-(2/n) * np.sum(np.subtract(y,predicted)))
    dtheta1 = (-(2/n) * np.sum(np.multiply(np.subtract(y,predicted),x)))

    theta0 = theta0 - (learningrate * dtheta0)
    theta1 = theta1 - (learningrate * dtheta1)
    
    histroy_theta0.append(theta0)
    histroy_theta1.append(theta1)
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(10,5))

# plot thetas over time

ax1.plot(histroy_theta0, label='$\\theta_{0}$', linestyle='--', color='tab:blue')
ax1.plot(histroy_theta1, label='$\\theta_{1}$', linestyle='-', color='tab:green')
# ax1.legend()
ax1.set_xlabel('Iterations'); ax1.set_ylabel('$\\theta$', color='tab:blue');
ax1.tick_params(axis='y', labelcolor='tab:blue')

# plot loss function over time
ax2 = ax1.twinx()
ax2.plot(loss, label='Loss function', color='tab:red')
ax2.set_title('Values of $\\theta$ and $J(\\theta)$ over iterations')
ax2.set_ylabel('Loss: $J(\\theta)$', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax2.grid()
ax1.grid()

# ax2.legend();
fig.legend();
```
