```python
'''
Language Modeling :

Train a model to predict the next word given a sequence of words.
Categorize the lanaguge modeling into two aspects:

    Auto - regressive language modeling
    Auto - encoding language modeling
    
Auto - regressive language modeling : 
    - categorize the auto regressive modeling as follows
        - Forward(left to right) prediction : 
                - model reads all the words from left to right up to the blank in order to make prediction.
                - Ex: paris is a beautiful _____.I love Paris
        - Backward(right to left) prediction:
                - model read all the words from right to left up to the blank in order to make prediction.
                - Ex: ____.I love Paris
    - Auto -regressive models are unidirectional in nature, meaning they read the sentence in only one direction.

Auto Encoding - Language Modeling:
    - Auto Encoding - Language Modeling takes advantages of both forward(left to right) and Backward(right to left) prediction.
     That is . it reads the sentence in both directions while making the prediction.
     - Auto encoding language model is bidirectional in nature.
     
Note : BERT is Auto Encoding language Model.

'''
```

```python
'''
Next Sentence Prediction (NSP):  
    - Another integresting strategy for used for trainig the the BERT model.
    - NSP is the binary classifcation task.
    - feed two sentences to BERT model and it has to predict whether the second sentence is the follow up (next sentence)
      of the first sentences.
      
Ex: 
    Sentence A: She Cooked pasta
    Sentence B: It was Delicious
    
    in the above Sentence B is a followup, that is the sentence following on from sentence A.
    So we label this sentence pair as "isNext",Indicating that sentence B follows on from sentence A.
    
    In the preceding sentence pair,sentence B is not follow Senetence B means We label this sentence pair as "notNext"
    
    Here we use Feedforward Layer with a Softmax Function which returns the probability.


'''
```

```python
'''
Subword Toknization:
    - It is Popularly used in Many state of the art model ,including GPT-3 and BERT.
    - It is Effective way to handle OOV(out of Vocabulary) words.
    
    
While building Vocabulary, we split Setnece into tokens and split the text present in the dataset by white space and all 
all unique words to vocabulary.

lets consider sentence : 'I enjoyed the game' and split sentence using whitespace
[I,enjoyed,the, game] now check whtehr all the words present in the vocabulaty and found that all words in the sentence
present in the vocabulary except 'enjoyed'. since word enjoyed not present in the vocabulary,we replace it with an unknow token.
<UNK>. so my final tokens looks like 
tokens = [I,<UNK>,the,game]

we can observe that althogh we have the word enjoy in the vocabulary, just becase of we din't have the exact word enjoyed,
It is marked as an unknow word with <UNK>

having a huge vocabulary causes memory and performance issues and it still connot handle the unseen words.

To handle above the issue, we can use of subword tokenization.

In Sub word tokenization we split words into subwords.
Suppose we split the word enjoyed into subwords [enjoy,ed] and after splitting the subwords, we add them to the vocabulary.

Now our vocabulary consists of the following

[game,the,I,play,walk,ed,enjoy]

tokens = [I,enjoy,##ed,the,game]

##ed indicates it is a subword and is preceded by another word.


Poular Subword Tokenization:
    - Byte pair encoding[BPE]
    - Byte-level byte pair encoding [BBPE]
    - WordPiece

'''
```

```python
'''
ALBERT :  A Lite version Of BERT 

One of the challenges with BERT model is it consists of millions of parameters.which makes harder to train and time consuming.
To overcome this.ALBERT was introduced.It is Lite version of BERT with fewer parameters compared to BERT.

***** Total 110 million parameters of BERT-base, the ALBERT model only has 31 million parameters while using the 
      same number of layers and 768 hidden units.
      
Techniques to Reduce the number of parameters in ALBERT
    - Cross-layer Parameter Sharing
    - Factorizing Embedding layer parameterization
    

    
Cross-layer Parameter Sharing :
    - We know BERTbase consists of 12 encoder layer.During training we learn the parameters of all the encoded layers.
    - In ALBERT instead of learning parameters for all layer, we only learn the parameters of the fist encoded layer, and then
      we just share the parameters of the first encoded layers with all the other encoded layers.
      
          **** Encoder consists of sublayers called Multihead Attention and FeedForward Layers.
        
        Several Options for Performing Cross-layer Parameter Sharing
            - All Shared : We share the parameters of all the sublayers of the first encoder with all the sublayers of the 
                           other encoder
            - Shared Feedforward Network : Only share the parameters of the feedforward network of the first encoder layer
                                           with the feedforward network other encoder layer
            - Shared Attention : Share only the parameters of Multi headed attention of the first encoder layer with the 
                                 Multi-head attention of other encoder layer.
         
         **** By defualt ALBERT uses the all shared option
         
Factorizing Embedding layer parameterization :



    - 

'''
```

```python
'''
RoBERT : It is essentilly BERT with the following chnages in the pre-traiing.
     - Use Dynamic Masking instead of Static Masking in the MLM tasks
     - Remove NSP (Next Sentence Predict)task and train using only MLM task
     - Train with lareg batch Size
     - use Byte-Level BPE [BBPE] as a tokenizer
     - Other Setting changes and Evaluated the model on Several dataset.
         - Experiments
                 1. SEGMENT-PAIR + NSP
                 2. SENTENCE_PAIR + NSP
                 3. FULL SENTENCES
                 4. DOC SENTENCES


'''

'''
what is Dynamic masking and Static masking?

    We know that Pre trained BERT using the MLM taks means we randomly mask 15% of the tokens and let the network predict the 
    masked tokens.
    The masking is done only once during the preprocessing step and we train the mdoel over several epoch to predict the
    same masked token it is called as Static Masking
    
    Coming to Dynamic Masking, we duplicate the sentence 10 times, next we radomly mask 15% of the tokens in all these 10 
    duplicate sentences, now we have 10 sentences with different tokens masked in each sentence.
    


'''
```

```python
'''
ELECTRA : Efficiently Learning and Encoder and Classfication Token Replacements

    - We learned that pre train BERT using the MLM and NSP taks. and in MLM 15% of the tokens are masked randomly.
    - In ELECTRA, model is pretrained using MLM with a task called Replaced Token Detection.
    
Replaced Token Detection : It is similar to MLM but instead of masking a token with the [MASK] token, we replace token
                           with different token and train the model to classify  whether the given tokens are actual or 
                           replaced tokens.
'''

'''
Why use Replaced Token Detection task instead of MLM task?
    - One problem with MLM is it uses [MASK] token will during the pre-training but the [MASK] tokenswill not be present
      during the fine tuning on the downstream task.this causes mismatch between pretraining and fine tuning.


'''
```
