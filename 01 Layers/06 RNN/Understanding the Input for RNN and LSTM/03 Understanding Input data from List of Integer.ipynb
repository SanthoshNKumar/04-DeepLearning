```python
X = list()
y = list()

X = [x+1 for x in range(20)]
Y = [y*15 for y in X]

print("X={0}".format(X))
print("Y={0}".format(Y))
```

```python
# Example 2

X = np.array([30])

X = X.reshape((1, 1, 1))

print(X)  # Sample =1 and Time step = 1 and Feature = 1
```

```python
# Example 3

X1 = [(x+1)*2 for x in range(25)]
X2 = [(x+1)*3 for x in range(25)]
Y = [x1*x2 for x1,x2 in zip(X1,X2)]

print("X1={0}".format(X1))
print("X2={0}".format(X2))
print("Y={0}".format(Y))

print("\n")

X = np.column_stack((X1, X2))
print(X)
X = np.array(X).reshape(25,1,2)

print("\n\n")
print("After Reshape")
print(X) # Sample = 25 and time step =1 and Feature = 2
```

```python
# Many to One Sequence

X = np.array([x+1 for x in range(45)])
print(X)

X = X.reshape(15,3,1)

print("\n")

print(X) # Sample = 15 and Time step =3 and Feature = 1
```

```python
# Many to One Sequence Problem with Multiple Features

X1 = np.array([x+3 for x in range(0, 135, 3)])
print("X1={0}".format(X1))

X2 = np.array([x+5 for x in range(0, 225, 5)])
print("X2={0}".format(X2))

print("\n")
X = np.column_stack((X1, X2))
print("X=\n{0}".format(X))

X = np.array(X).reshape(15,3,2)  

print("\n")

print(X) # Sample = 15 and Time step =3 and Features =2
```
