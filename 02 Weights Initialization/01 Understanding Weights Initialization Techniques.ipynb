```python
'''
Note :

    - fan-in refers to the number of units in the previous layer
    - fan-out refers to the number of units in the subsequent layer

'''
```

```python
'''

Standard Normal initialization — this approach samples each weight from a normal distribution with low deviation

Lecun initialization — these initializations produce weights that are randomly selected numbers multiplied with 
                       the variance 1/fan-in
    
Xavier initialization : Randomly generated weight is multiplied by variance 2/(fan-in + fan-out)
    
He initialization : Randomly generated weights and multiplies them by 2/fan-in

'''
```

```python
'''

weights range

1. Xavier initialization(Uniform Distribution)
2. Xavier initialization(Normal Distribution)

1. He initialization(Uniform Distribution)
2. He initialization(Normal Distribution)

'''s
```

```python
#02 Uniform Initialization
    # weights range = -(1.0 / sqrt(fan_in)), (1.0 / sqrt(fan_in))

# example of the xavier weight initialization
from math import sqrt
from numpy import mean
from numpy.random import rand

# number of nodes in the previous layer
fan_in = 10

# calculate the range for the weights
lower, upper = -(1.0 / sqrt(fan_in)), (1.0 / sqrt(fan_in))

# generate random numbers
numbers = rand(1000)

# scale to the desired range
scaled = lower + numbers * (upper - lower)

# summarize
print(lower, upper)
print(scaled.min(), scaled.max())
print(scaled.mean(), scaled.std())


# kernel_initializer=kernel_initializers.RandomUniform(minval=-0.05,maxval=0.05)
```
