```python
#01 : RandomNormal :Initializer that generates tensors with a normal distribution.
initializer = tf.keras.initializers.RandomNormal(mean=0., stddev=1.)

initializer(shape=(3,6))
```

```python
#02 : RandomUniform
# Initializer that generates tensors with a uniform distribution.

initializer = tf.keras.initializers.random_uniform(minval=0., maxval=1.)

initializer(shape=(3,6))
```

```python
#03 : TruncatedNormal

initializer = tf.keras.initializers.TruncatedNormal()

initializer(shape=(3,6))
```

```python
#04 : Zeros

initializer = tf.keras.initializers.zeros()

initializer(shape=(3,6))
```

```python
#05 : Ones

initializer = tf.keras.initializers.ones()

initializer(shape=(3,6))
```

```python
#06 : GlorotNormal

initializer = tf.keras.initializers.glorot_normal()

initializer(shape=(3,6))
```

```python
#07 : GlorotUniform

initializer = tf.keras.initializers.glorot_uniform()

initializer(shape=(3,6))
```

```python
#08 : Identity

initializer = tf.keras.initializers.identity()

initializer(shape=(4,8))
```

```python
#08 : Orthogonal

initializer = tf.keras.initializers.Orthogonal()

initializer(shape=(4,8))
```

```python
#09 : Constant
initializer = tf.keras.initializers.Constant(5)

initializer(shape=(4,8))
```

```python
#10 : VarianceScaling

initializer = tf.keras.initializers.VarianceScaling()

initializer(shape=(4,8))
```

```python
#11 : he_normal
initializer = tf.keras.initializers.he_normal()

initializer(shape=(4,8))
```

```python
#11 : he_uniform
initializer = tf.keras.initializers.he_uniform()

initializer(shape=(4,8))
```

```python
# How to set custom weights in keras using NumPy array


input_shape=(28,28,1)
model = tf.keras.Sequential(
    [
        tf.keras.Input(shape=input_shape,name='input'),
        tf.keras.layers.Conv2D(32, kernel_size=(3, 3),kernel_initializer=tf.keras.initializers.zeros(),
                               activation="relu",use_bias=True,name='conv1'),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(10, activation="relu",use_bias=True,bias_initializer='zeros',name='dense1'),
        tf.keras.layers.Dense(1, activation="softmax",name='output'),
    ]
)
```
