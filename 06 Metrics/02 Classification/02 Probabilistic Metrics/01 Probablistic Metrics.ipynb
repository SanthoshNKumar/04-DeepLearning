```python
'''

What is Log Loss?
    Log Loss is the most important classification metric based on probabilities.

Note : For a given problem a lower the log loss means better predictions

Mathematical interpretation:
    Log Loss is the negative average of the log of corrected predicted probabilities for each instance.
    
'''
```

```python
'''

Steps to implement Log Loss
    1. To find corrected probabilities.
    2. Take a log of corrected probabilities.
    3. Take the negative average of the values we get in the 2nd step.
'''
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"ID":['ID6','ID1','ID7','ID8','ID2','ID3','ID4','ID5'],
              "Actual":[1,1,1,0,0,1,1,0],
              "Predicted Probabilities":[0.94,0.9,0.78,0.56,0.51,0.47,0.32,0.1]})

df
```

```python
'''
What are the corrected probabilities?
    By default, the output of the logistics regression model is the probability of the sample being positive.
'''
```

```python
# Corrected Probabilities = (1 - Predicted Probabilities) where Actual = 0

df['Corrected Probabilities'] = [x if y ==1 else (1-x) for x,y in zip(df['Predicted Probabilities'],df['Actual']) ]
```

```python
df
```

```python
import numpy as np

df['Log'] = [np.log10(x) for x in df['Corrected Probabilities']]

df
```

```python
logloss = df['Log'].sum() / 8

logloss
```
