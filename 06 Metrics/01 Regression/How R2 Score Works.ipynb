```python
# R2 Score
    - coefficient of determination
    - R2 score = Variance explained by the model  / Total variation
    
    - 0% corresponds to a model that does not explain the variability of the response data around its mean
        - The mean of the dependent variable helps to predict the dependent variable and also the regression model. 
        
    - 100% corresponds to a model that explains the variability of the response variable around its mean. 
    - R2 score = 1 - (SSE/SST)
        - SSE : Sum of Squares Residual Error
        - SST : Sum of Squared Total Error
            
    - Note :
        - The best possible score is 1 which is obtained when the predicted values are the same as the actual values.
        - R2 score of baseline model is 0.
        - During the worse cases, R2 score can even be negative
```

```python
What is Goodness of Fit:
    - Deviations between the observed values and the predicted values of the linear model are small and unbiased, 
      the model has a well-fit data or goodness of fit
```

```python
# Case 1 : R2 Score is Undefined
#    - All the data is 5.0.
#    - SST is 0 because no variance in the dependent Varibale(All data points are same)
#    -  = 1- (89/0) = Error : eroDivisionError: division by zero
# when  all the dapapoints are same SST become zero

y= [5,5,5,5,5,5]

y_bar = np.mean(y)
SST = np.sum([(y_t - y_bar)**2 for y_t in y])
print("Denominator SST value:",SST)
```
