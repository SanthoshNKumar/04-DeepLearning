```python
'''
Learning Curve to identify Overfitting and Underfitting in Machine Learning

Learning curves show the relationship between training set size and your chosen evaluation metric (e.g. RMSE, accuracy, etc.)
on your training and validation sets.

They can be an extremely useful tool when diagnosing your model performance, as they can tell you whether your model is 
suffering from bias or variance

'''
```

```python
'''

Image 1:

If your learning curves look like this, it means your model is suffering from high bias. 
Both the training and validation (or cross-validation) error is high and it doesn’t seem to improve with more training examples.
The fact that your model is performing similarly bad for both the training and validation sets suggests that 
the model is underfitting the data and therefore has high bias

'''
```

```python
'''

Image 2:

If your learning curves look like this, your model might have a high-variance problem. 
In this chart, the validation error is much higher than the training error, which suggests that you are overfitting the data.
'''
```
