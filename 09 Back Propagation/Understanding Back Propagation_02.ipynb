```python
import numpy 
x1=0.1
x2=0.4

target = 0.7
learning_rate = 0.01

w1=numpy.random.rand()
w2=numpy.random.rand()

print("Initial W : ", w1, w2)

predicted_output = []
network_error = []
epoch = []
```

```python
import numpy as np

def sigmoid(sop):
    return 1.0/(1+numpy.exp(-1*sop))
```

```python
for i in range(1000):
    
    epoch.append(i)
    
    y = w1*x1 + w2*x2
    
    predicted = 1 / (1 + np.exp(-y))
    
    err = np.power(predicted-target, 2)
    
    predicted_output.append(predicted)
    network_error.append(err)
    
    g1 = 2*(predicted-target) # Derivate of Error functiom

    g2 = sigmoid(y)*(1.0-sigmoid(y)) # derivate of Sigmoid Function
    
    g3w1= x1 # Derivate Weights
    g3w2= x2 # Derivate of weights
    
    gradw1 = g3w1*g2*g1   # Gradient Descent Descent
    gradw2 = g3w2*g2*g1
    
    w1 = w1 - learning_rate*gradw1
    w2 = w2 - learning_rate*gradw2
```

```python
import matplotlib.pyplot as plt

plt.plot(epoch,network_error)
```
