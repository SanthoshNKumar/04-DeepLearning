```python
'''

Formaula for finding output of the layers in the CNN

1. Layer without padding 
 ((n - f) / s ) + 1  *  ((n - f) / s ) + 1 

      n = Image size (ex : 227 )
      f = filter size 
      s = stride

2. Layer with padding
  ((n + 2p - f) / s ) + 1  *  ((n + 2p - f) / s ) + 1 
  
  n = Image size (ex : 227 )
  p = padding
  f = filter size 
  s = stride
    
'''
```

```python
'''

AlexNet
    - Alexnet won the Imagenet large-scale visual recognition challenge in 2012
    - AlexNet architecture consists of 5 convolutional layers, 3 max-pooling layers, 2 normalization layers, 
      2 fully connected layers, and 1 softmax layer.
    - Each convolutional layer consists of convolutional filters and a nonlinear activation function ReLU
    - The pooling layers are used to perform max pooling. 
    - Input size is fixed due to the presence of fully connected layers.
    - AlexNet overall has 60 million parameters.
    - Batch size of 128 
    - SGD Momentum as learning algorithm 
    - Heavy Data Augmentation with things like flipping, jittering, cropping, color normalization, etc. 
    - AlexNet was trained on a GTX 580 GPU with only 3 GB of memory which couldnt fit the entire network.
      So the network was split across 2 GPUs, with half of the neurons(feature maps) on each GPU. 
      
'''
```

```python
'''
VGG 16 : 
    - VGG : Visual Geometry Group
    - It is also called the OxfordNet model, named after the Visual Geometry Group from Oxford.
    - Number 16 refers that it has a total of 16 layers that has some weights.
    - It Only has Conv and pooling layers in it.
    - always use a 3 x 3 Kernel for convolution.
    - 2x2 size of the max pool.
    - has a total of about 138 million parameters.
    - Trained on ImageNet data
    - It has an accuracy of 92.7%.
    - it has one more version of it Vgg 19, a total of 19 layers with weights.

'''
```

```python
'''
Architecture of 'VGG'

    Total Layer : 16

    Input : 224*224*3
    
    13 Convolution Layer
    1 Flattern Layer
    2 Dense Layer
    
    VGG 16 Having 6 Blocks
        1. 1st Block hvaing 2 CONV Layer and 1 Max Pooling
        2. 2nd Block hvaing 2 CONV Layer and 1 Max Pooling
        3. 3rd Block hvaing 3 CONV Layer and 1 Max Pooling
        4. 4th Block hvaing 3 CONV Layer and 1 Max Pooling
        5. 5th Block hvaing 3 CONV Layer and 1 Max Pooling
        6. 6th Block Having 2 Flattern and 2 Dence Layer

    All the Convolution layer
        - Filter size : 3X3
        - Stride = 1
        - Padding = same
        
    Max pooling 
        - Filter size :2x2 
        - stride =2 

    Output : (1,1000)
'''
```

```python
'''
ResNet:(Residual Network)
    - Won 1st place in the ILSVRC 2015 classification competition with top-5 error rate of 3.57% 
    - Won the 1st place in ILSVRC and COCO 2015 competition in ImageNet Detection, ImageNet localization, Coco detection 
      and Coco segmentation
    - ResNet, which was proposed in 2015 by researchers at Microsoft Research introduced a new architecture called Residual 
      Network.
    - Residual Network some where address vanishing/exploding Gradient problem.
    - Resnet use the concept of Skip connection. using this 'skip connection' or 'Identity mapping' we can addess the 
      problem of Vanishing Gradinet Problem.


    Skip Connection : Normally Convolution layer are stcked one after the After , but in Resnet layers are stcked one after 
    other also add original input to the output of Convolution.

    Traditional : Yhat = F(x)

    Skip connection = Yhat = F(x) + x

    Here main goal of to make F(x) = 0
    so we can achive Yhat = x
    
    
    RenNet 50 : 
        - ResNet50 is a variant of ResNet model which has 48 Convolution layers along with 1 MaxPool and 1 Average Pool layer
        
    Variants of ResNets
        - ResNeXt
        - DenseNet : Densely Connected CNN
        
    ResNets:
        - ResNet V1
        - ResNet V2
            - ResNet50 
            - ResNet101
            - ResNet152
            - ResNet50V2
            - ResNet101V2
            - ResNet152V2

'''
```
